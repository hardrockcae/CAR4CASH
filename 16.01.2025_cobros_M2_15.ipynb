```python

import matplotlib.pyplot as plt

# Ruta al archivo .xls
file_path = r"C:\Users\hp\Documents\archivo_completo_nuevo.xls"

# Leer el archivo Excel
df_cobros = pd.read_excel(file_path, engine='xlrd')

# Convertir 'Fecha del principio de contrato' a formato de fecha si no está ya en ese formato
df_cobros['Fecha del principio de contrato'] = pd.to_datetime(df_cobros['Fecha del principio de contrato'], errors='coerce')

# Filtrar las filas con fechas válidas y valores numéricos en 'Vencimiento'
df_cobros = df_cobros.dropna(subset=['Fecha del principio de contrato', 'Vencimiento'])

# Ordenar por 'Fecha del principio de contrato' para una curva clara
df_cobros = df_cobros.sort_values(by='Fecha del principio de contrato')

# Crear el gráfico
plt.figure(figsize=(12, 6))
plt.plot(df_cobros['Fecha del principio de contrato'], df_cobros['Vencimiento'], marker='o', linestyle='-', label='Vencimiento vs. Fecha del principio')

# Etiquetas y título
plt.title('Comparativa entre Vencimiento y Fecha del principio de contrato', fontsize=14)
plt.xlabel('Fecha del principio de contrato', fontsize=12)
plt.ylabel('Vencimiento', fontsize=12)
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ruta al archivo
file_path = r"C:\Users\hp\Desktop\CAR4CASH\exportData-2025-01-16.xls"

# Leer el archivo Excel
df_cobros_completo = pd.read_excel(file_path, engine='xlrd')

# Convertir 'Fecha del principio de contrato' a formato de fecha
df_cobros_completo['Fecha del principio de contrato'] = pd.to_datetime(df_cobros_completo['Fecha del principio de contrato'], errors='coerce')

# Crear una nueva columna con solo Mes/Año
df_cobros_completo['Mes/Año de inicio'] = df_cobros_completo['Fecha del principio de contrato'].dt.to_period('M').astype(str)

# Filtrar filas válidas para 'Vencimiento'
df_cobros_completo = df_cobros_completo.dropna(subset=['Mes/Año de inicio', 'Vencimiento'])

# Agrupar por 'Mes/Año de inicio' y 'Vencimiento', contando la cantidad de clientes
df_agrupado = df_cobros_completo.groupby(['Mes/Año de inicio', 'Vencimiento']).size().reset_index(name='Cantidad de clientes')

# Pivotar los datos para preparar el gráfico
df_pivot = df_agrupado.pivot(index='Mes/Año de inicio', columns='Vencimiento', values='Cantidad de clientes').fillna(0)

# Crear el gráfico de barras apiladas
df_pivot.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab10')

# Personalizar el gráfico
plt.title('Comparativa: Clientes por Mes/Año de inicio y Vencimiento', fontsize=16)
plt.xlabel('Mes/Año de inicio', fontsize=12)
plt.ylabel('Cantidad de clientes', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Vencimiento', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
