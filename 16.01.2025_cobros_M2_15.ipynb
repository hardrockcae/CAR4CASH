```python
import pandas as pd

# Configurar pandas para mostrar todo el DataFrame
pd.set_option('display.max_rows', None)  # Mostrar todas las filas
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
pd.set_option('display.width', None)  # Ajustar el ancho para mostrar todo

# Ruta al archivo .xls
file_path = r"C:\Users\hp\Documents\archivo_completo_nuevo.xls"

# Leer el archivo Excel
df_cobros = pd.read_excel(file_path, engine='xlrd')

# Imprimir todo el DataFrame
print("Contenido completo del DataFrame df_cobros:")
print(df_cobros)
```

```python


# Leer el archivo Excel
df_cobros = pd.read_excel(file_path, engine='xlrd')

# Filtrar los datos según las condiciones
df_filtrado = df_cobros[
    (df_cobros['Vencimiento'] >= -2) & 
    (df_cobros['Vencimiento'] <= 15) & 
    (df_cobros['Firma del contrato'] == 'Sí')
]

# Seleccionar las columnas necesarias
df_resultado = df_filtrado[
    ['ID', 'Teléfono', 'Correo electrónico', 'Fecha del principio de contrato', 
     'Plazo mensual', 'Vencimiento', 'Nombre', 'Empresa/Apellido', 
     'Código postal', 'Ciudad', 'Marca', 'Modelo']
]

# Imprimir el resultado
print("Resultados filtrados:")
print(df_resultado)

# Opcional: Exportar los resultados a un archivo Excel
output_path = r"C:\Users\hp\Documents\df_cobros_filtrado.xlsx"
df_resultado.to_excel(output_path, index=False)
print(f"\nArchivo filtrado guardado en: {output_path}")

```
